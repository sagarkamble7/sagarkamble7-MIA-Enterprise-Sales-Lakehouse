### **DYNAMIC INPUTS**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
dbutils.widgets.text("file_name", "")
file_name = dbutils.widgets.get("file_name")
```

### **DATA READING**

```python
# df = spark.read.format("parquet")\
#     .load(f"abfss://bronze@customersprojectete.dfs.core.windows.net/{file_name}")

# df.count()
```

```python
df = spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "parquet")\
    .option("cloudFiles.schemaLocation", f"abfss://bronze@customersprojectete.dfs.core.windows.net/checkpoint_{file_name}")\
    .load(f"abfss://source@customersprojectete.dfs.core.windows.net/{file_name}")
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", f"abfss://bronze@customersprojectete.dfs.core.windows.net/checkpoint_{file_name}")\
    .option("path", f"abfss://bronze@customersprojectete.dfs.core.windows.net/{file_name}")\
    .trigger(once=True)\
    .start()
```
