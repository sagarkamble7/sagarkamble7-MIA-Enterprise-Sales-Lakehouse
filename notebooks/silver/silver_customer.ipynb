```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
from pyspark.sql.types import *
```

```python
df = spark.read.format("parquet")\
    .load("abfss://bronze@customersprojectete.dfs.core.windows.net/customers")
```

```python
df.limit(5).display()
```

```python
df = df.withColumn("domains", split(col("email"),"@")[1])
df.display()
```

```python
df = df.withColumn("fullname", concat(col("first_name"), lit(" "), col("last_name")))
df = df.drop("first_name", "last_name")
df.limit(5).display()
```

```python
df = df.drop("_rescued_data")
df.limit(1).display()
```

```python
df.write.mode("overwrite")\
    .format("delta")\
    .save("abfss://silver@customersprojectete.dfs.core.windows.net/customers")
```

```python
%sql
create schema databricks_cata.silver;
```

```python
%sql
create table if not exists databricks_cata.silver.customers_silver
using delta
location "abfss://silver@customersprojectete.dfs.core.windows.net/customers"
```
